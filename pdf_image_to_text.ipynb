```python
import pytesseract
from PIL import Image
from pdf2image import convert_from_path
import os

# Set the Tesseract executable path

def extract_text_from_image(image_path):
    try:
        print(f"Processing image: {image_path}")
        image = Image.open(image_path)
        text = pytesseract.image_to_string(image)
        return text
    except Exception as e:
        print(f"Error reading image {image_path}: {e}")
        return None

def extract_text_from_pdf(pdf_path):
    try:
        print(f"Converting PDF to images: {pdf_path}")
        # Convert PDF to a list of images
        images = convert_from_path(pdf_path)
        print(f"Number of pages converted: {len(images)}")

        # Extract text from each image
        extracted_texts = []
        for i, image in enumerate(images):
            print(f"Extracting text from page {i + 1}")
            text = pytesseract.image_to_string(image)
            extracted_texts.append(text)

        # Combine text from all pages
        combined_text = "\n".join(extracted_texts)
        return combined_text
    except Exception as e:
        print(f"Error reading PDF {pdf_path}: {e}")
        return None

# Define the directory containing your files
directory = r'D:\DL'  # Adjust your path here

# List all files in the directory
files = os.listdir(directory)

# Dictionary to store extracted text
extracted_texts = {}

# Extract text from each file and store it in the dictionary
for file in files:
    file_path = os.path.join(directory, file)
    file_lower = file.lower()
    if file_lower.endswith(('.png', '.jpg', '.jpeg', '.tiff', '.bmp', '.gif')):
        text = extract_text_from_image(file_path)
    elif file_lower.endswith('.pdf'):
        text = extract_text_from_pdf(file_path)
    else:
        print(f"Skipping file (not an image or PDF): {file}")
        continue

    if text:
        extracted_texts[file] = text

# Display the extracted text
for file, text in extracted_texts.items():
    print(f"Text from {file}:\n{text}\n")
```
